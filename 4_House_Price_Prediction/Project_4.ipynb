### Step 3 :- Data Preprocessing

#### Check Missing Values

```python
# Display missing values
house_data.isnull().sum()
```

#### Fill Missing Values

```python
# Fill missing values
house_data = house_data.fillna(

    house_data.median(

        numeric_only = True))
```

#### Check Duplicate Records

```python
# Display duplicate records
print(f'Duplicate Records: {house_data.duplicated().sum()}')
```

#### Remove Duplicate Records

```python
# Remove duplicate records
house_data = house_data.drop_duplicates()

# Display duplicate records
print(f'Duplicate Records: {house_data.duplicated().sum()}')
```

#### Check Data Types

```python
# Display data types
house_data.dtypes
```

#### Separate Features and Target

> Replace `Price` with your actual target column name if it is different.

```python
# Separate features and target
features = house_data.drop('Price', axis = 1)

target = house_data['Price']
```

#### Split Training and Testing Data

```python
# Split dataset
X_train, X_test, y_train, y_test = train_test_split(

    features,

    target,

    test_size = 0.2,

    random_state = 42)
```

#### Display Dataset Shape

```python
# Display dataset shape
print(f'Training Features: {X_train.shape}')

print(f'Testing Features: {X_test.shape}')

print(f'Training Target: {y_train.shape}')

print(f'Testing Target: {y_test.shape}')
```

### Step 4 :- Model Training

#### Create Linear Regression Model

```python
# Create linear regression model
linear_regression_model = LinearRegression()
```

#### Train Model

```python
# Train model
linear_regression_model.fit(

    X_train,

    y_train)
```

#### Display Model Coefficients

```python
# Display model coefficients
coefficients = pd.DataFrame({

    'Feature': X_train.columns,

    'Coefficient': linear_regression_model.coef_})

coefficients
```

#### Sort Coefficients

```python
coefficients = coefficients.sort_values(

    by = 'Coefficient',

    ascending = False)

coefficients
```

#### Display Intercept

```python
# Display intercept
print(f'Intercept: {linear_regression_model.intercept_:.2f}')
```

#### Display Number of Features

```python
# Display number of features
print(f'Total Features: {X_train.shape[1]}')
```

#### Display Training Score

```python
# Display training score
training_score = linear_regression_model.score(

    X_train,

    y_train)

print(f'Training R² Score: {training_score:.4f}')
```

### Step 5 :- Model Evaluation

#### Predict Test Data

```python
# Predict test data
predicted_price = linear_regression_model.predict(

    X_test)
```

#### Compare Actual vs Predicted Prices

```python
# Compare actual and predicted prices
prediction_results = pd.DataFrame({

    'Actual Price': y_test,

    'Predicted Price': predicted_price})

prediction_results.head(10)
```

#### Calculate Mean Absolute Error (MAE)

```python
# Calculate MAE
mae = mean_absolute_error(

    y_test,

    predicted_price)

print(f'MAE: {mae:.2f}')
```

#### Calculate Mean Squared Error (MSE)

```python
# Calculate MSE
mse = mean_squared_error(

    y_test,

    predicted_price)

print(f'MSE: {mse:.2f}')
```

#### Calculate Root Mean Squared Error (RMSE)

```python
# Calculate RMSE
rmse = np.sqrt(

    mse)

print(f'RMSE: {rmse:.2f}')
```

#### Calculate R² Score

```python
# Calculate R² score
r2 = r2_score(

    y_test,

    predicted_price)

print(f'R² Score: {r2:.4f}')
```

#### Display Evaluation Summary

```python
# Display evaluation summary
evaluation_results = pd.DataFrame({

    'Metric': [

        'MAE',

        'MSE',

        'RMSE',

        'R² Score'],

    'Value': [

        mae,

        mse,

        rmse,

        r2]})

evaluation_results
```
